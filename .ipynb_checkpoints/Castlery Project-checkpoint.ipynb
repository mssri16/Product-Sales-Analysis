# Castlery Products' Sales Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams.update({'font.size': 22})
```

**Load the file from local system and read the individual sheets**

```python
#load the file
file = pd.ExcelFile('Marketing_Analytics_Intern_Assignment_(1).xlsx')
```

```python
#read and load the individual sheets into dataframe
product_sales_data = pd.read_excel(file,'data_1')
product_lead_time = pd.read_excel(file,'data_2')

# product_sales_data.info()
# product_lead_time.info()
```

**Data Cleaning & Pre-Processing**

```python
###change product column from float64 to object(string) for better readability

product_sales_data['product'] = product_sales_data['product'].astype('object')
product_lead_time['product'] = product_lead_time['product'].astype('object')

### recheck the types of the columns in the data

product_sales_data.info()

product_lead_time.info()
```

```python
### check if there are any duplicate rows in the given data

product_sales_data[product_sales_data.duplicated()]
product_lead_time[product_lead_time.duplicated()]
```

```python
### check if there are any duplicate product_id in the given data

product_id_df = product_sales_data['product']
product_id_df[product_id_df.duplicated()]
```

```python
### join data_1 & data_2 on product_id and date

overall_product_data = pd.merge(product_sales_data, product_lead_time, left_on=['product','date'], right_on =['product','date'])
overall_product_data.head()
```

```python
### data anomaly where purchase is 1 and view is 0

overall_product_data.drop(overall_product_data[overall_product_data['purchases']>overall_product_data['views']].index, inplace=True)
overall_product_data[overall_product_data['purchases']>overall_product_data['views']]
```

```python
### introduce conversionrate column to the dataframe; conversionrate = purchases/views

overall_product_data['conversionrate'] = overall_product_data['purchases'] / overall_product_data['views']
overall_product_data.head()
```

```python
### introduce unitprice column for all the products

overall_product_data['unitprice'] = overall_product_data['sales'] / overall_product_data['purchases']

overall_product_data.head()
```

```python
### replace null category values with N/A

overall_product_data['category'].fillna('N/A',inplace=True)
overall_product_data[overall_product_data['category'].isna()]
```

```python
### convert date column into string values and store it separately

overall_product_data['converteddate'] = overall_product_data['date'].astype(str)
```

**Data Analysis (Lead Time Vs Conversion Rate)**

```python
### check the overall correlation between leadtime and conversionrate
overall_product_data['leadtime'].corr(overall_product_data['conversionrate'])
```

```python
### scatter plot leadtime Vs conversionrate

plt.scatter('leadtime','conversionrate',data=overall_product_data)
plt.xlabel('Lead Time')
plt.ylabel('Conversion Rate')
plt.title('Lead Time Vs Conversion Rate')
plt.show()
```

```python
### scatter plot leadtime vs conversionrate only for the purchased products to compare with the old scatter plot

purchased_products_df = overall_product_data[overall_product_data['purchases']>0]
plt.scatter('leadtime','conversionrate',data=purchased_products_df)
# plt.plot(overall_product_data['leadtime'],np.log(overall_product_data['conversionrate']),'r')
plt.xlabel('Lead Time')
plt.ylabel('Conversion Rate')
plt.title('Lead Time Vs Conversion Rate')
plt.show()
```

```python
### construct a dataframe with average conversionrate since high variance in conversionrate for every leadtime

avg_data = overall_product_data.groupby(['leadtime'])['conversionrate'].mean().reset_index()
avg_data = avg_data.rename(columns={'conversionrate':'avg_conversion_rate'})
# std_data = overall_product_data.groupby(['leadtime'])['conversionrate'].std().reset_index()
# avg_data['std'] = std_data['conversionrate']
avg_data
```

**Linear (Logarithmic) Regression - Average trend line**

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

y = avg_data['avg_conversion_rate'].values
x = avg_data['leadtime'].values

z = np.log(x, where=x > 0)

z = z.reshape(z.shape[0],1)
y = y.reshape(y.shape[0],1)

model = LinearRegression()
model.fit(z, y)

coef = model.coef_
intercept = model.intercept_
```

```python
### plot the average curve line for the above constructed data (average conversion rate vs lead time)

from sklearn.metrics import r2_score
plt.figure()
plt.plot(avg_data['leadtime'], avg_data['avg_conversion_rate'], 'ko', label="Original Noised Data")
x= avg_data['leadtime']
plt.plot(x, intercept+(coef[0]*np.log(x, where=x>0)), color='red')
plt.show()

# r2_score(avg_data['avg_conversion_rate'],intercept+(coef[0]*np.log(x, where=x>0)))
```
